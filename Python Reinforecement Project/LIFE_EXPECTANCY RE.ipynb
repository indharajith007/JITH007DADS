# **3. HYPOTHESIS TESTING**

### ^^ **ONE-SAMPLE T-TEST**

-SCIPY.STATS--

```python
import scipy.stats as stats
```

```python
life_exp = df['Life expectancy '].dropna()   # drop NaN values
t_stat, p_value = stats.ttest_1samp(life_exp, 70)
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Mean life expectancy is significantly different from 70")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference from 70")
```

### INSIGHTS

* THE TEST CHECKS IF THE MEAN LIFE EXPECTANCY IS 70.

* **P-VALUE** IS **MUCH LESS THAN 0.05** â†’ **REJECT THE NULL HYPOTHESIS.**

* MEAN **LIFE EXPECTANCY** IS **SIGNIFICANTLY DIFFERENT** FROM **70.**

### ^^ **INDEPENDENT TWO-SAMPLE T-TEST**

```python
developed = df[df['Status'] == 'Developed']['Life expectancy '].dropna()
developing = df[df['Status'] == 'Developing']['Life expectancy '].dropna()
t_stat, p_value = stats.ttest_ind(developed, developing)
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Significant difference between Developed & Developing countries' life expectancy")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference")
```

### INSIGHTS
- THIS CODE IS A **TWO-SAMPLE T-TEST** COMPARING LIFE EXPECTANCY BETWEEN **DEVELOPED** AND **DEVELOPING** COUNTRIES.
    
- **P-VALUE** IS FAR BELOW **0.05** â†’ **REJECT THE NULL HYPOTHESIS**.
  
- THERE IS A **SIGNIFICANT DIFFERENCE** BETWEEN THE LIFE EXPECTANCY OF **DEVELOPED** AND **DEVELOPING** COUNTRIES

### ^^ **PAIRED T-TEST**

```python
life_exp = df['Life expectancy '].dropna()
schooling = df['Schooling'].dropna()
paired_data = df[['Life expectancy ', 'Schooling']].dropna()
life_exp = paired_data['Life expectancy ']
schooling = paired_data['Schooling']
t_stat, p_value = stats.ttest_rel(life_exp, schooling)
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Significant difference between Life expectancy & Schooling")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference")
```

### INSIGHTS

* THIS CODE IS A **PAIRED T-TEST** COMPARING **LIFE EXPECTANCY** AND **SCHOOLING** OF THE **SAME COUNTRY**  

* **P-VALUE** IS **EXTREMELY SMALL (<0.05)** â†’ **REJECT NULL HYPOTHESIS**  

* THERE IS A **SIGNIFICANT DIFFERENCE** BETWEEN **LIFE EXPECTANCY** AND **SCHOOLING**  


### ^^ **ONE-SAMPLE Z-TEST**

```python
from statsmodels.stats.weightstats import ztest
```

**1. TO CHECK IF THE AVERAGE LIFE EXPECTANCY = 70**

```python
life_exp = df['Life expectancy '].dropna()
z_stat, p_value = ztest(life_exp, value=70)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Mean life expectancy is significantly different from 70")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference from 70")
```

### **INSIGHTS**

* MEAN **LIFE EXPECTANCY** IS SLIGHTLY DIFFERENT FROM **70**  

* **P-VALUE** IS MUCH LESS THAN **0.05** â†’ DIFFERENCE IS SIGNIFICANT  

* THERE IS STRONG EVIDENCE THAT **LIFE EXPECTANCY** DIFFERS FROM **70**  


**2. TO CHECK IF THE MEAN SCHOOLING = 12 YEARS**

```python
schooling = df['Schooling'].dropna()
z_stat, p_value = ztest(schooling, value=12)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Mean Schooling is significantly different from 12")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference from 12")
```

### **INSIGHTS**

* MEAN **SCHOOLING YEARS** IS SLIGHTLY DIFFERENT FROM **12**  

* **P-VALUE** IS GREATER THAN **0.05** â†’ DIFFERENCE IS NOT SIGNIFICANT  

* NO STRONG EVIDENCE THAT **SCHOOLING** DIFFERS FROM **12**  


### ^^**TWO-SAMPEL Z-TEST** 

**1. COMPARE LIFE EXPECTANCY FOR DEVELOPED VS DEVELOPING**

```python
developed = df[df['Status'] == 'Developed']['Life expectancy '].dropna()
developing = df[df['Status'] == 'Developing']['Life expectancy '].dropna()
z_stat, p_value = ztest(developed, developing)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Significant difference between Developed and Developing countries' Life Expectancy")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference")
```

### ^ **INSIGHTS**

* DEVELOPED AND DEVELOPING COUNTRIES SHOW NOTICEABLY DIFFERENT AVERAGE **LIFE EXPECTANCY**

* THE HIGH Z-STATISTIC **(e.g., 12.45)** SHOWS THE DIFFERENCE IS STRONG AND UNLIKELY DUE TO CHANCE

* P-VALUE **(e.g., 2.1e-20)** IS FAR BELOW **0.05**, MEANING THE DIFFERENCE IS STATISTICALLY SIGNIFICANT

* THE NULL HYPOTHESIS (**NO DIFFERENCE**) IS REJECTED WITH HIGH CONFIDENCE


**2. COMPARE LIFE EXPECTANCY FOR HIGH VS LOW GDP**

```python
median_gdp = df['GDP'].median()
high_gdp = df[df['GDP'] >= median_gdp]['Life expectancy '].dropna()
low_gdp = df[df['GDP'] < median_gdp]['Life expectancy '].dropna()
z_stat, p_value = ztest(high_gdp, low_gdp)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Significant difference between High and Low GDP countries' Life Expectancy")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference")
```

## ^ **INSIGHTS**

* COUNTRIES WITH **HIGH GDP** HAVE CLEARLY HIGHER **LIFE EXPECTANCY** COMPARED TO LOW GDP COUNTRI
    
* THE LARGE Z-STATISTIC (e.g., **10.3**) SHOWS THE DIFFERENCE IS STRONG AND NOT DUE TO CHA
  E  
* P-VALUE (e.g., **3.4e-15**) IS FAR BELOW 0.05, INDICATING A HIGHLY SIGNIFICANT DIFFER
  CE  
* NULL HYPOTHESIS (**NO DIFFERENCE**) IS REJECTED WITH HIGH CONFIDENCE  


**2. COMPARE LIFE EXPECTANCY FOR HIGH VS LOW SCHOOLING COUNTRIES**

```python
median_schooling = df['Schooling'].median()
high_schooling = df[df['Schooling'] >= median_schooling]['Life expectancy '].dropna()
low_schooling = df[df['Schooling'] < median_schooling]['Life expectancy '].dropna()
z_stat, p_value = ztest(high_schooling, low_schooling)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject Hâ‚€ â†’ Significant difference between High and Low Schooling countries' Life Expectancy")
else:
    print("Fail to reject Hâ‚€ â†’ No significant difference")
```

## ^ **INSIGHTS**

* COUNTRIES WITH **HIGHER SCHOOLING YEARS** HAVE NOTICEABLY HIGHER **LIFE EXPECTANCY
    
* Z-STATISTIC (e.g., **8.7**) SHOWS THE EFFECT IS STRONG AND RELIA
  E  
* P-VALUE (e.g., **2.5e-12**) IS WELL BELOW 0.05, CONFIRMING SIGNIFICANT DIFFER

# **MULTIVARIET ANALYSIS**

### -- **CORRELATION HEATMAP**

--THIS **CORRELATION HEATMAP**  VISUALLY SHOW THE STRONGLY PAIRS OF VARIABLES ARE RELATED

^^THIS **CORRELATION HEATMAP VISUALIZATION** TO SHOW THE COUNT OF THE **LIFE EXPECTANCY DATASET**

```python
data.columns = data.columns.str.strip()
life_data = data[['Life expectancy', 'GDP', 'Schooling', 'BMI']]
corr = life_data.corr()
plt.figure(figsize=(8,6))
ax = sns.heatmap(
    corr, 
    cmap="coolwarm", 
    fmt=".2f", 
    linewidths=0.5, 
    square=True, 
    cbar_kws={"shrink": 0.8},
    annot=False   
)
for i in range(len(corr)):
    for j in range(len(corr)):
        value = corr.iloc[i, j]
        text_color = "white" if abs(value) > 0.5 else "black"  # adaptive color
        ax.text(
            j+0.5, i+0.5, f"{value:.2f}",
            ha="center", va="center",
            color=text_color, fontsize=12, fontweight="bold"
        )
plt.title("Correlation Heatmap - Life Expectancy Dataset", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

### **INSIGHTS**

- **LIFE EXPECTANCY VS SCHOOLING** â†’ STRONG POSITIVE CORRELATION  
   MORE YEARS OF SCHOOLING ARE LINKED WITH HIGHER LIFE EXPECTANCY  

- **LIFE EXPECTANCY VS GDP** â†’ MODERATE POSITIVE CORRELATION

- **LIFE EXPECTANCY VS BMI** â†’ WEAK / VERY LOW CORRELATION
